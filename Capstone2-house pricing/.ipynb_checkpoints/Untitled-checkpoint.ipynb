```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno
```

```python
train = pd.read_csv('train.csv')
```

```python
train.head().T
```

```python
train.info()
```

```python
train.shape
```

```python
# check out the missing value
missing = pd.concat([train.isnull().sum(),100*train.isnull().mean()],axis = 1)
missing.columns = ['# of NaN','Percentage']
missing.sort_values('Percentage',ascending = False,inplace = True)
```

```python
missing.head(25)
```

```python
# The list that the missing value is meaningful
None_list = ['PoolQC','MiscFeature','Alley','Fence','FireplaceQu','GarageFinish','GarageCond','GarageType','GarageType','GarageQual','BsmtFinType2','BsmtFinType1','BsmtExposure','BsmtCond','BsmtQual','MasVnrType']
```

```python
# Replace the missing value with None

for feature in None_list:
    train[feature].fillna('None', inplace = True)

# Check if the missing values have been replaced    
train.isna().sum().sort_values(ascending = False)
```

```python
train.loc[:,~train.columns.isin(None_list)]
```

```python
train.describe().T
```

```python
# Drop the ID column
train.drop('Id',axis =1, inplace = True)
train.describe().T
```

```python
# Some values doesn't make sense, convert it to string
train['MSSubClass'] = train['MSSubClass'].astype('string')
```

```python
# check out the remaining missing value
missing2 = pd.concat([train.isnull().sum(),100*train.isnull().mean()],axis = 1)
missing2.columns = ['# of NaN','Percentage']
missing2.sort_values('Percentage',ascending = False,inplace = True)
```

```python
missing2.head(10)
```

```python
# Replace the missing value in Electrical with the most oftenly used company in the same neighborhood
train.groupby('Neighborhood').Electrical.value_counts()
```

```python
train[train.Electrical.isna()].Neighborhood
```

```python
missing_electrical = train[train.Neighborhood == train[train.Electrical.isna()].Neighborhood].Electrical.mode()
```

```python
train.Electrical.fillna(missing_electrical[0], inplace = True)
```

```python
train.info()
```

```python
# Fill missing value in MasVnrArea and GarageYrBlt with 0 since it means the house doesn't have mansonry venner or garage
train.update(train[['MasVnrArea','GarageYrBlt']].fillna(0))
```

```python
train.info()
```

```python
train.plot.scatter(x='LotFrontage',y='SalePrice')
plt.show()
```

```python
# Replace the missing valuein LotFrontage with the mean value in its own neighborhood
train.groupby('Neighborhood').LotFrontage.mean()
train['LotFrontage'] = train.groupby('Neighborhood')['LotFrontage'].transform(lambda x: x.fillna(x.mean()))
```

```python
train.describe()
```
